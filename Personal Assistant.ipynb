```python
import speech_recognition as sr
from time import ctime
import time
import os
from gtts import gTTS
from googlesearch import search
import sys
import webbrowser as wb
import speak2


chrome_path="C:\Program Files (x86)\Microsoft\Edge\Application\msedge.exe %s"

def speak(audioString):
    print(audioString)
    tts = gTTS(text=audioString, lang='en')
    lang = 'en'
    speak2.tts(audioString, lang)

def recordAudio():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("...")
        audio = r.listen(source)
        print(audio)

    data = ""
    try:
        data = r.recognize_google(audio)
        print("You said: " + data)
    except sr.UnknownValueError:
        print("I could not understand what you said")
    except sr.RequestError as e:
        print("Could not request results from Speech Recognition service; {0}".format(e))
 
    return data
 
def myspeech(data):
    
    if "hello Alexa" or "Alexa" in data:
        speak("Hello! How are you?")        

    if ("i am fine" or "i am good") in data:
        speak("Good to see that :)")
        
    if "how are you" in data:
        speak("I am fine")
        
    if ("what time is it" or "what's the time" or "what is the time") in data:
        speak(ctime())
 
    if "where is" in data:
        data = data.split(" ")
        location = data[2]
        speak("Hold on Shivam, I will show you where " + location + " is.")
        os.system("https://www.google.nl/maps/place/" + location + "/&amp;")
        
    if "search for"in data:
        data = data.split(" ")
        searchdata = data[2]
        speak("Ok Shivam, I'm Searching " + searchdata + " for you.")
        f_text ='https://www.google.com/search?q='+searchdata
        wb.get(chrome_path).open(f_text)

    if "what"in data:
        data = data.split(" ")
        searchdata = ' '.join(map(str, data))
        speak("Ok Shivam, I'm Searching " + searchdata + " for you.")
        f_text ='https://www.google.com/search?q='+searchdata
        wb.get(chrome_path).open(f_text)

    if "play" in data:
        data = data.split(" ")
        searchdata = data[2]
        speak("do you want to play the song on YouTube, Saavn or Windows Media Player?")

        data1 = recordAudio()
        playsong(data1,searchdata)

    if "shopping list" in data:
        speak("You wish to add or delete something or you want to view your shopping list?")
        data2 = recordAudio()
        shoplist(data2)
  
    if "thank you" in data:
        speak("You're Welcome!")
        sys.exit()

def shoplist(data2):
    a = "E:\Personal Assistant\Shopping List.txt"
    if "add" in data2:
        data2 = data2.split(" ")
        item = data2[1]

        if(os.path.exists(a)):
            f=open(a, "a")
            f.writelines(item)
            
        else:
            f=open(a, "w")
            f.writelines(item)
            
            speak("do you want to add something else?")
            data3 = recordAudio()
            if "yes" in data3:
                data2 = recordAudio()
                shoplist(data2)
            f.close()
            
    if "delete" in data2:
        data2 = data2.split(" ")
        item = data2[1]
        if(os.path.exists(a)):
            keyword=item
            filename=a
            print('generator started')
            f = open(filename, 'r')

            try:
                for line in f:
                    if keyword in line:
                        f.write(" ")
            except:
                print("Iteration exception")
            f.close()
            f=open(a, "r")
            s= f.read()
            print(s)
            f.close()
            
        else:
            print("No Shopping List Yet")
            speak("No shopping list yet")
            
    if "view" in data2:
        f=open(a, "r")
        s= f.read()
        print(s)
        f.close()

    speak("do you want to add something else?")
    data3 = recordAudio()
    if "yes" in data3:
        data4 = recordAudio()
        shoplist(data4)
    f.close()
    
def playsong(data1,searchdata):
    if "YouTube" in data1:
        speak("Ok Shivam, Playing on Youtube.")
        f_text ='https://www.youtube.com/results?search_query='+searchdata
        wb.get(chrome_path).open(f_text)
        start()
    elif "Saavn" in data1:
        speak("Ok Shivam, Playing on Saavn.")
        f_text ='https://www.saavn.com/search/'+searchdata
        wb.get(chrome_path).open(f_text)
        start()

    elif "Windows media player" in data1:
        speak("Ok Shivam, Playing on Windows Music Player.")
        os.startfile('C:\ProgramData\Microsoft\Windows\Start Menu\Programs\Accessories\Windows Media Player.lnk')
        start()
                    
def start():
    data = recordAudio()

    if data == "Alexa":
        speak("Hi Shivam, what can I do for you?")
        while 1:
            data = recordAudio()
            print(data)
            myspeech(data)
    else:
        print("....")
        start()
start()
```
